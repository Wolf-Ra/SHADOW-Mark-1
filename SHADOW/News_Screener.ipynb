```python
# Function to fetch and categorize latest news
def fetch_news():
    gn = GNews(language='en', country='IN')
    news_list = []
    
    # Fetch top news
    news = gn.get_top_news()
    
    for article in news:
        title = article['title']
        description = article.get('description', '')
        link = article['url']
        pub_date = article['published date']
        full_text = f"{title} {description}"
        
        category_assigned = 'Uncategorized'

        # Append data for this article
        news_list.append({
            'category': category_assigned,
            'title': title,
            'description': description,
            'link': link,
            'published_at': pub_date,
        })
    return news_list
```

```python
News_data = pd.DataFrame(columns=['category', 'title', 'description', 'link', 'published_at'])
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()

    # Tokenize the text
    words = word_tokenize(text.lower())
    # Remove stopwords and lemmatize
    words = [lemmatizer.lemmatize(word) for word in words if word.isalnum() and word not in stop_words]
    return words
```

```python
# Extract trends using TF-IDF
def extract_trends(word_list):
    documents = word_list
    vectorizer = TfidfVectorizer(max_features=10)
    X = vectorizer.fit_transform(documents)
    return vectorizer.get_feature_names_out()
```

```python
list_after_nltk = []
for sentence in news_list_nltk:
    News_des = preprocess_text(sentence)
    News_des = extract_trends(News_des)
    list_after_nltk.append(News_des)
list_after_nltk
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment(text):
    if text:
        sentiment = analyzer.polarity_scores(text)
        return sentiment['compound']  # Sentiment score ranging from -1 to 1
    return 0
```

```python
df_Sentiment['date'] = pd.to_datetime(df_Sentiment['date'])
df_Sentiment
```

```python
def Category_wise_sentiment(df_Sentiment):
    sentiment = []
    sentiment_list_category = []
    category = ["Automobile","Banking and Finance","IT","Pharmaceuticals","Energy","Consumer Goods","Metals and Mining","Real Estate","Telecommunication","Infrastructure","Textiles","Cement and Construction Materials","Media and Entertainment","Tourism","Retail"]
    for i in range(len(df_Sentiment)):
        title = df_Sentiment['title'].iloc[i]
        des = df_Sentiment['description'].iloc[i]
        full_text = f"{title} {des}"
        sent = analyze_sentiment(full_text)
        sentiment.append(sent)
    df_Sentiment['Sentiment'] = sentiment
    for item in category:
        tot_sentiment = 0
        temp_df = pd.DataFrame()
        temp_df = df_Sentiment.loc[(df_Sentiment['category'] == item)]
        for j in range(len(temp_df)):
            tot_sentiment = tot_sentiment + temp_df['Sentiment'].iloc[j]
        sentiment_list_category.append(tot_sentiment)
    Sentiment_data = pd.DataFrame(columns=category)
    Sentiment_data.loc[len(Sentiment_data.index)] = sentiment_list_category
    Sentiment_data['date'] = df_Sentiment['date']
    return Sentiment_data
```

```python
Sentiment_data = pd.DataFrame()
Sentiment_data = Category_wise_sentiment(df_Sentiment)
Sentiment_data
```

```python
from gliner import GLiNER
```

```python
# Initialize GLiNER Model for Named Entity Recognition
model = GLiNER.from_pretrained("urchade/gliner_base")
```

```python
def Find_Companies(df_Sentiment):
    Company_list = []
    business_labels = ["COMPANY", "ORGANIZATION", "PRODUCT"]
    for i in range(len(df_Sentiment)):
        title = df_Sentiment['title'].iloc[i]
        entities = model.predict_entities(title,business_labels)
        company_name = [entity['text'] for entity in entities]
        Company_list.append(company_name)
    df_Sentiment['Company'] = Company_list
    return df_Sentiment
```

```python
df_Sentiment = Find_Companies(df_Sentiment)
df_Sentiment
```

```python
def Company_contribution(df_Sentiment,Sentiment_data,category):
    Company_contribution_list = []
    company_df = df_Sentiment.loc[(df_Sentiment['category'] == category)]
    for i in range(len(company_df)):
        Company_contribution_list.append({
            "date" : company_df['date'].iloc[i],
            "Company" : company_df['Company'].iloc[i],
            "Contribution" : company_df['Sentiment'].iloc[i],
            "Total_Sentiment" : Sentiment_data[category].iloc[len(Sentiment_data)-1],
            "Ratio of Contribution" : company_df['Sentiment'].iloc[i]/Sentiment_data[category].iloc[len(Sentiment_data)-1]*100,
            "News" : company_df['title'].iloc[i],
            "Description" : company_df['description'].iloc[i]
        })
    Company_pie = pd.DataFrame(Company_contribution_list)
    return Company_pie
```

```python
category = ["Automobile","Banking and Finance","IT","Pharmaceuticals","Energy","Consumer Goods","Metals and Mining","Real Estate","Telecommunication","Infrastructure","Textiles","Cement and Construction Materials","Media and Entertainment","Tourism","Retail"]
for item in category:
    print(" ")
    print("For ",item," :")
    company_pie = Company_contribution(df_Sentiment,Sentiment_data,item)
    company_pie.to_excel(item+'.xlsx',index=False)
    print(company_pie)
```
